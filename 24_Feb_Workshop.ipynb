```python
import pandas as pd
import numpy as np
data = {
    'ID': ['1', '2', '2', '4', '5'], # Duplicate ID + string type
    'Name': ['Amit', 'Riya', 'Riya', 'Kiran', None], # Missing value
    'Marks': [80, 75, 75, 75, "absent"], # Missing values
    'Grade': ['A', 'B', 'B', None, 'B'] # Missing + categorial
}
df = pd.DataFrame (data)
print (df)
print ("\n", df.isnull()) # Returns True for missing values
print ("\n", df.isnull().sum()) # Returns sum of missing values
print ("\n", df.notnull()) # Returns Non-null values
print ("\n", df.dropna(axis = 1)) # Drop columns with missing value
print ("\n", df.dropna()) # Drop rows with missing value
print ("\n", df.dropna (how = 'all')) # Drop rows where all values are missing
print ("\n", df.dropna (thresh = 2)) # Drop rows with at least 2 missing values
# print ("\n", df.dropna (subset = ["Name"])) # Creating copy
# df = df.dropna (subset = ["Name"], inplace = True)
print ("\n", df)
# print ("\n", df['Marks'].fillna(df['Marks'].mean()))
print ("\n", df['Marks'].fillna(0))
# print ("\n", df['Marks'].fillna(df['Marks'].median()))
print ("\n", df['Marks'].fillna(df['Marks'].mode()[0]))
print ("\n", df.duplicated())
print ("\n", df.drop_duplicates())
print ("\n", df['ID'].astype (int))
print ("\n", pd.to_numeric (df ['Marks'], errors = 'coerce'))
print ("\n", df ["Marks"])
```
